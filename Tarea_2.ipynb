```python
########### Solución 6
np.random.seed(55555)
point_xy = np.random.randint(0,100,[30,2])
print(point_xy)
```

```python
########### Solución 6 estudiante 1
min_red = [np.minimum.reduce(list(map(lambda i: point_xy[i][0], range(len(point_xy))))), 
           np.minimum.reduce(list(map(lambda i: point_xy[i][1], range(len(point_xy)))))]
min_red
```
